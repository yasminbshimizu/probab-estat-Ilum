## 9.1 Goblin

Considerando que a variável aleatória $X$ apresenta a *pdf*
$$
f(x) =\begin{cases}
c(1-x²) & -1 < x < 1 \\
0 & \textrm{caso contrário}
\end{cases}
$$

#### 1. Encontre o valor de $c$ .
Pela expressão $f(x)$ da *pdf*, podemos concluir que a probabilidade do espaço amostral será $0$ para $x < -1$ e $x > 1$.

Sabendo que a probabilidade do espaço amostral de uma *pdf* deve ser igual a $1$:

$$
\int_{-1}^{1} f(x) \, \mathrm{d}x = 1 
$$

$$
\int_{-1}^{1} c(1 - x²) \, \mathrm{d}x = 1
$$

$$
c\int_{-1}^{1} (1 - x²) \, \mathrm{d}x = 1 
$$

$$
c (x - \frac{x³}{3}) \bigg|_{-1}^1 = 1
$$

$$
c (1 - \frac{1}{3} - (-1 - \frac{-1}{3}))= 1
$$

$$
c (1 - \frac{1}{3} +1 - \frac{1}{3})= 1
$$

$$
\frac{4c}{3}= 1
$$

$$
\therefore c = \frac{3}{4}
$$

Sendo assim, concluímos que $c = \frac{3}{4}$ , logo, $f(x) =\begin{cases}
\frac{3}{4}(1-x²) & -1 < x < 1 \\
0 & \textrm{caso contrário}
\end{cases}
$ .

#### 2. Encontre a forma analítica da *cdf* de $X$.

Sabemos que, por definição, uma *cdf* é dada por:


$$
F(x) = P(X \leq x) = \int_{-\infty}^{x} f(x) \, \mathrm{d}x.
$$

Sendo assim, aplicando a *pdf* da nossa valiável aleatória, temos, considerando $x\leq-1$:

$$
F(x) = P(X \leq x) = \int_{-\infty}^{x} f(x) \, \mathrm{d}x
$$

$$
F(x) = P(X \leq x) = \int_{-\infty}^{x} 0 \, \mathrm{d}x 
$$

$$
\therefore F(x) = P(X \leq x) = 0
$$

Para $-1<x<1$ :

$$
F(x) = P(X \leq x) = \int_{-\infty}^{x} f(x) \, \mathrm{d}x
$$

$$
F(x) = P(X \leq x) = \int_{-\infty}^{-1} 0 \, \mathrm{d}x + \int_{-1}^{x} \frac{3}{4}(1-x²) \mathrm{d}x
$$

$$
F(x) = P(X \leq x) = \int_{-1}^{x} \frac{3}{4}(1-x²) \mathrm{d}x
$$

$$
F(x) = P(X \leq x) =\frac{3}{4} (x-\frac{x³}{3}) \bigg|_{-1}^x
$$

$$
F(x) = P(X \leq x) =\frac{3}{4} (x-\frac{x³}{3}) - \frac{3}{4} (-1-\frac{-1}{3})
$$

$$
\therefore F(x) = P(X \leq x) =\frac{1}{2}+\frac{3x}{4}-\frac{x³}{4}
$$

Por fim, para $x\geq1$ :

$$
F(x) = P(X \leq x) = \int_{-\infty}^{x} f(x) \, \mathrm{d}x
$$

$$
F(x) = P(X \leq x) = \int_{-\infty}^{-1} 0 \, \mathrm{d}x + \int_{-1}^{1} \frac{3}{4}(1-x²) \mathrm{d}x + \int_{1}^{x} 0 \, \mathrm{d}x
$$

$$
F(x) = P(X \leq x) =\int_{-1}^{1} \frac{3}{4}(1-x²) \mathrm{d}x 
$$


$$
F(x) = P(X \leq x) =\frac{3}{4} (x-\frac{x³}{3}) \bigg|_{-1}^1
$$

$$
F(x) = P(X \leq x) =\frac{3}{4} (1-\frac{1}{3}) - \frac{3}{4} (-1-\frac{-1}{3})
$$

$$
\therefore F(x) = P(X \leq x) =1
$$

Sendo assim, concluímos que a *cdf* de $X$ é 
$F(x) = P(X \leq x) = 
\begin{cases}
0 &  x \leq -1 \\
\frac{1}{2}+\frac{3x}{4}-\frac{x³}{4} & -1<x<1 \\
1 & x \geq -1
\end{cases}
$ .

#### 3. Calcule $P(|X| > 0,5)$.

Podemos dividir o problema em duas partes: $P(|X| > 0,5)=P(X < -0,5) + P(X > 0,5)$.

É possível, ainda, representar tal problema de uma outra forma: $P(|X| > 0,5)=1 - P(-0,5 \leq X \leq 0,5)$

Sendo assim, sabendo que a probabilidade de um valor estar entre $a$ e $b$ é de $P(a \leq X \leq b) = \int_{a}^{b} f(x) \, \mathrm{d}x$, temos que:

$$
P(-0,5 \leq X \leq 0,5) = \int_{-0,5}^{0,5} f(x) \, \mathrm{d}x
$$

$$
P(-0,5 \leq X \leq 0,5) = \int_{-0,5}^{0,5} \frac{3}{4}(1-x²) \, \mathrm{d}x
$$

$$
P(-0,5 \leq X \leq 0,5) = \frac{3}{4} (x-\frac{x³}{3}) \bigg|_{-0,5}^{0,5}
$$

$$
P(-0,5 \leq X \leq 0,5) = \frac{3}{4} (\frac{1}{2}-\frac{1}{2³\times3}- (\frac{-1}{2}-\frac{-1}{2³\times3}))
$$

$$
\therefore P(-0,5 \leq X \leq 0,5) = \frac{11}{16}
$$

Sendo assim, podemos concluir que , como $P(|X| > 0,5)=1 - P(-0,5 \leq X \leq 0,5)$, $P(|X| > 0,5)= 1 - \frac{11}{16} \therefore P(|X| > 0,5) = \frac{5}{16}$.

#### 4. Calcule $E[X]$.

Sabendo que o valor esperado em uma distribuição contínua é dado por

$$
E[X] = \int_{-\infty}^{\infty} x \cdot f(x) \, \mathrm{d}x
$$

para encontrar $E[X]$ basta substituirmos o $f(x)$ e resolvermos a integral.

$$
E[X] = \int_{-\infty}^{-1} 0 \, \mathrm{d}x + \int_{-1}^{1} x\frac{3}{4}(1-x²) \mathrm{d}x + \int_{1}^{x} 0 \, \mathrm{d}x
$$

$$
E[X] =  \frac{3}{4}\int_{-1}^{1} x(1-x²) \mathrm{d}x
$$

$$
E[X] =  \frac{3}{4}\int x(1-u) \frac{\mathrm{d}u}{2x} , u = x² 
$$

$$
E[X] =  \frac{3}{8}\int (1-u) \mathrm{d}u , u = x² 
$$

$$
E[X] = \frac{3}{8} (u - \frac{u²}{2}), u = x²
$$

$$
E[X] = \frac{3}{8} (x² - \frac{(x²)^2}{2})\bigg|_{-1}^{1}
$$

$$
E[X] = \frac{3}{8} (1 - \frac{1}{2}) - \frac{3}{8} (-1 - \frac{1}{2})
$$

$$
E[X] = \frac{3}{8} (1 - \frac{1}{2}) + \frac{3}{8} (1 + \frac{1}{2})
$$

$$
\therefore E[X] = \frac{3}{4}
$$

Desse modo, podemos concluir que o valor esperado da variável aleatória $X$ é $E[X] = \frac{3}{4}$. É interessante notar que $E[X] = c$, que também é intercepto da *pdf* de $X$, sendo que 0 é a mediana de $f(x)$.

#### 5. Calcule ${Var}(X)$.

Sabemos que a variância de uma variável aleatória $X$ é dada por:
$$
\mathrm{Var}(X) = E[X^2] - (E[X])^2
$$

Então:

$$
\mathrm{Var}(X) = E[X^2] - (E[X])^2
$$ 

$$
\mathrm{Var}(X) = - \frac{3²}{4²} + \int_{-1}^{1} x²\frac{3}{4}(1-x²) \mathrm{d}x
$$ 

$$
\mathrm{Var}(X) = - \frac{3²}{4²} + \frac{3}{4}\int_{-1}^{1} (x²-x^4) \mathrm{d}x
$$ 

$$
\mathrm{Var}(X) = - \frac{9}{16} + \frac{3}{4} (\frac{x³}{3}-\frac{x^5}{5}) \bigg|_{-1}^1
$$ 

$$
\mathrm{Var}(X) = - \frac{9}{16} + \frac{3}{4} (\frac{1}{3}-\frac{1}{5}) - \frac{3}{4} (\frac{-1}{3}-\frac{-1}{5})
$$

$$
\mathrm{Var}(X) = - \frac{9}{16} + \frac{3}{4} (\frac{1}{3}-\frac{1}{5}) + \frac{3}{4} (\frac{1}{3}-\frac{1}{5})
$$

$$
\mathrm{Var}(X) = - \frac{9}{16} + \frac{1}{5}
$$

$$
\therefore {Var}(X) = - \frac{29}{80}
$$

Desse modo, concluímos que a variância de $X$ é ${Var}(X) = - \frac{29}{80}$.
