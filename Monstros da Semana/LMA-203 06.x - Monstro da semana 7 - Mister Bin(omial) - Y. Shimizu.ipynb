## Introdução

Considerando uma partida de volêi hipotética, observaremos a probabilidade de uma jogadora hipotética Fulana marcar um *ace* hipotético ao realizar seu saque hipotético. Para isso, analisamos a função massa de probabilidade (*pmf*) para diferentes números de repetição de saques hipotética num treino hipotético, plotando os resultados obtidos graficamente.

Para isso, implementamos a função `binom.pmf(k, n, p)` do `scipy` para calcular as probabilidades da *pmf*, a qual usa como base a expressão:
$$
f(k)=\begin{cases}
\binom{n}{k}p^{k}(1-p)^{n-k} & k=0,1,2,\ldots,n & 0\leq p \leq 1\\
0 & \textrm{caso contrário}
\end{cases}
$$

onde $n$ é o número de repetições, $k$ é o número de sucessos e $p$ é a probabilidade de sucesso.

## Metodologia

Primeiramente, importamos as bibliotecas necessárias:

```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt
```

Então, definimos a probabilidade $p$ de Fulana marcar um ponto de saque:

```python
PROB_p = np.random.uniform (0.2,0.8)

print(f"A probabilidade p de Fulana marcar um ace é de {PROB_p}")
```

Feito isso, basta calcular as probabilidades da *pmf* para $n$ repetições de saque, com diversos valores de $n$. Aqui, escolhemos analisar a *pmf* para cinco série diferentes: $n_1=6$, $n_2=20$, $n_3=60$, $n_4=99$ e $n_5=300$.

### **$n_1 = 6$**

```python
n1 = 6

num_sucessos1 = range(n1+1)

prob1 =binom.pmf(num_sucessos1,n1, PROB_p)

print(f"A massa de distribuição de probabilidade calculada é:{ prob1}.")

plt.bar(num_sucessos1, prob1, color="rebeccapurple")

plt.title(f"pmf:  p={round(PROB_p*100,1)}% & n = {n1} ")
plt.xlabel("Número de sucessos")
plt.ylabel("$p(x)$");
```

## Discussões

Calculando as *pmf*'s para cada uma das cinco situações simuladas, para $n_1=6$, $n_2=24$, $n_3=60$, $n_4=102$, e $n_5=300$ com $p=73.5$, foi possível perceber um padrão interessante. O gráfico de distribuição de probabilidades apresenta um comportamento semelhante a uma curva gaussiana, com o seu ponto máximo diretamente relacionado à probabilidade inicial $p=73.5$, independente do número de ensaios. Assim, o número de sucessos com maior probabilidade de acontecer será $x$, tal que $\frac{x}{n}\approx p$, sendo que, nesse caso, a probabilidade de haver 0 sucessos em qualquer uma das situações definidas é praticamente nula, com uma alta densidade de probabilidade concentrada na segunda metade do intervalo definido. No caso do primeiro gráfico, por $n1$ ainda ser relativamente pequeno, vemos uma distribuição mais uniforme, com altas probabilidades em $k=4$ e $k=5$; para o segundo gráfico, com $n_2=24$, vemos que as probabilidades começam a subir em $k=10$, sendo que para $k=25$ também não há uma probabilidade mínima; já para o terceiro caso, as probabilidades são altas entre $k=30$ e $k=55$; para o quarto, $k=60$ e $k=90$; e para o quinto *pmf*, entre $k=200$ e $k=250$. É possível analisar também que quanto maior o número de ensaios, menor a largura do pico de probabilidades.

## Conclusões
Neste trabalho, concluímos que a *pmf* de um distribuição binomial apresenta um comportamento gaussiano, com o deslocamento do pico associado diretamente à probabilidade $p$ de sucesso, que modod que o número de sucessos com maior probabilidade de acontecer será $x$, tal que $\frac{x}{n}\approx p$. Foi possível observar também que a largura do pico diminui com o aumento de ensaios, havendo uma concentração de probabilidade de certos resultados numa faixa específica do intervalo determinado por $n$.
