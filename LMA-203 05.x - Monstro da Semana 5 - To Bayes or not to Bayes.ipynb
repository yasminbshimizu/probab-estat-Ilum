## Introdução
No trabalho de pesquisa de duas teorias da literatura, concluímos que o evento 1 tem maior confiabilidade. Sendo assim, foram feitas 3 sequências de 100 experimentos buscando calcular a probabilidade a posteriori do evento 1 acontecer através do Teorema de Bayes. Nessas sequências, obtivemos como probabilidade inicial:
* 1. Probabilidade Inicial: 20% | Acurácia: 99%
* 2. Probabilidade Inicial: 20% | Acurácia: 50%
* 3. Probabilidade Inicial: 99% | Acurácia: 5%

Além de testar a probabilidade final após um experimento, fizemos uma série de 10 experimentos recursivos para cada caso, visando obter resultados mais conclusivos para a hipótese.

## Metodologia
Para o cálculo da probabilidade, criamos duas funções na linguagem computacional Python para a Lei da Probabilidade Total e para o o Teorema de Bayes. Além disso, uma terceira função foi feita para o cálculo de n Teoremas de Bayes sucessivos, de modo que, com o resultado de cada probabilidade inicial, foi plotado um gráfico de pontos com tais.

```python
import matplotlib.pyplot as plt
```

```python
def prob_total(inicial,acuracia):
    evento = acuracia*inicial + (1-acuracia)*(1-inicial)
    return evento
```

```python
def bayes(inicial,acuracia):
    final = inicial*acuracia/prob_total(inicial,acuracia)
    return final
```

```python
i1 = 0.2
a1 = 0.99

E1 = round(bayes(i1,a1),4)
print(E1)
```

```python
i2 = 0.2
a2 = 0.5

E2 = round(bayes(i2,a2),4)
print(E2)
```

```python
i3 = 0.99
a3 = 0.05

E3 = round(bayes(i3,a3),4)
print(E3)
```

```python
def bayes_sucessivos(inicial, acuracia, n):
    in_list = [inicial]
    for e in range(n):
        inicial = bayes(inicial, acuracia)
        in_list.append(inicial)
    return in_list
        
in_list1 = bayes_sucessivos(i1,a1,10)
print(in_list1)

plt.plot(range(11), in_list1, marker="o")
plt.title("Valores de probabilidade inicial do Evento 1")
```

```python
in_list2 = bayes_sucessivos(i2,a2,10)
print(in_list2)

plt.plot(range(11), in_list2, marker="o")
plt.title("Valores de probabilidade inicial do Evento 1")
```

```python
in_list3 = bayes_sucessivos(i3,a3,10)
print(in_list3)

plt.plot(range(11), in_list3, marker="o")
plt.title("Valores de probabilidade inicial do Evento 1")
```

## Discussões
Através do algoritmo desenvolvido, as probabilidades finais obtidas foram:
* 1. Probabilidade Inicial: 20% | Acurácia: 99% | Probabilidade Final: 96,12%
* 2. Probabilidade Inicial: 20% | Acurácia: 50% | Probabilidade Final: 20%
* 3. Probabilidade Inicial: 99% | Acurácia: 05% | Probabilidade Final: 83,9%

Com as aplicações sucessivas do teorema de Bayes, percebemos que o melhor experimento para o evento 1 é o primeiro caso, já que, em apenas 2 iterações, a probabilidade de acontecer aumentou para quase 100%. Com o mesmo procedimento, o evento 2 manteve probabilidade constante ao longo das iterações e o evento 3 diminuiu sua probabilidade muito rápido. Isso se deve às acurácias de cada experimento: para o caso 1, a acurácia a1>0,5 leva à confirmação do resultado do experimento após sucessivos experimentos; para o 2, a acurácia a2=0,5 não muda a probabilidade do evento ao longo dos experimentos; e  no evento 3, a acurácia a3<0,5 leva à probabilidade 0% do evento acontecer após sucessivos experimentos.

## Conclusões
Desse modo, concluímos que, as probabilidades finais de cada caso de nosso experimento foram 96,12% , 20% e 83%, o que fortalece a hipótese de que o evento 1 é de maior confiabilidade. Além disso, concluímos também que, o caso 1 é o ideal no âmbito do evento 1, visto sua apurácia elevadíssima, de modo que testes sucessivos levam, na maioria dos casos, à confirmação do evento (probabilidade máxima).
